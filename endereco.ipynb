### Críticas

```python
class Critica:
    """
    ['ID', 'ENDERECO', 'BAIRRO', 'CIDADE', 'UF', 'CEP']
    """

    def __init__(self, banco_dados: pd.DataFrame):

        self.banco_dados = banco_dados.astype(str)

    @staticmethod
    def _aplicar_regras(valor: str, dict_regras: dict) -> str:

        for padrao, rotulo in dict_regras.items():
            if re.match(padrao, valor):
                return rotulo
        return 'OUTROS'
    
    @staticmethod
    def _aplicar_regras_bad_word(valor: str, dict_regras: dict) -> str:

        for padrao, rotulo in dict_regras.items():
            if re.match(padrao, valor):
                return rotulo
        return 'OK'

    def _extrair_regex(self, nome_coluna_origem: str, nome_coluna_destino: str, dict_regras: dict) -> pd.DataFrame:

        self.banco_dados[nome_coluna_destino] = (
            self.banco_dados[nome_coluna_origem]
            .fillna('')
            .astype(str)
            .apply(lambda v: self._aplicar_regras(v, dict_regras))
        )
        return self.banco_dados

    def critica_cep(self) -> pd.DataFrame:

        self.banco_dados['CEP'] = self.banco_dados['CEP'].str.strip()

        dict_cep = {
            r'^\s*$': 'BRANCO',
            r'^\d{8}$': 'OK',
            r'^(\d)\1+$': 'NÚMERO REPETIDO',
            r'^(?:01234567|12345678|23456789|98765432|87654321|76543210)$': 'NÚMERO SEQUENCIAL',
            r'^(?:012|123|234|345|456|567|678|789|890|987|876|765|654|543|432|321|210)$': 'NÚMERO SEQUENCIAL CURTO'
        }
        return self._extrair_regex('CEP', 'CRITICA_CEP', dict_cep)

    def critica_uf(self) -> pd.DataFrame:

        self.banco_dados['UF'] = self.banco_dados['UF'].str.strip().str.upper()

        dict_uf = {
            r'^\s*$': 'BRANCO',
            r'^[A-Z]{2}$': 'OK'
        }
        return self._extrair_regex('UF', 'CRITICA_UF', dict_uf)

    def _normalizar(self, nome: str):

        # tratar col antes de aplicar reg

        self.banco_dados[nome] = (
            self.banco_dados[nome]
            .str.strip()
            .str.normalize('NFKD')#testar token
            .str.encode(encoding='ascii', errors='ignore') # sem o ignore pode dar erro
            .str.decode('utf-8')
            .str.upper()
        )
        #unicodedata.normalize("NFKD", texto).encode("ASCII", "ignore").decode("utf-8")

    def critica_cidade(self) -> pd.DataFrame:
        
        self._normalizar('CIDADE')

        dict_cidade = {
            r'^\s*$': 'BRANCO',
            r'^[A-Z\s]+$': 'OK',
            r'^[0-9]+$': 'NÚMERO',
            r'^[!@#$%^&*(),.?":{}|<>]+$': 'CARACTERES ESPECIAIS',
            r'^[A-Z\s]*[0-9]+[A-Z\s]*$': 'NÚMERO NO MEIO DO TEXTO',
            r'^[A-Z\s]*[!@#$%^&*(),.?":{}|<>]+[A-Z\s]*$': 'CARACTERES ESPECIAIS NO MEIO DO TEXTO'
        }
        return self._extrair_regex('CIDADE', 'CRITICA_CIDADE', dict_cidade)

    def critica_bairro(self) -> pd.DataFrame:
        self._normalizar('BAIRRO')
        dict_bairro = {
            r'^\s*$': 'BRANCO',
            r'^[A-Z\s]+$': 'OK',
            r'^[0-9]+$': 'NÚMERO',
            r'^[!@#$%^&*(),.?":{}|<>]+$': 'CARACTERES ESPECIAIS',
            r'^[A-Z\s]*[0-9]+[A-Z\s]*$': 'NÚMERO NO MEIO DO TEXTO',
            r'^[A-Z\s]*[!@#$%^&*(),.?":{}|<>]+[A-Z\s]*$': 'CARACTERES ESPECIAIS NO MEIO DO TEXTO'
        }
        return self._extrair_regex('BAIRRO', 'CRITICA_BAIRRO', dict_bairro)


    def critica_bad_word(self, bad_words : list, coluna_base : str) -> pd.DataFrame:

        dict_bad_word = {fr'.*\b{word}\b.*': 'BAD WORD' for word in bad_words}

        nome_col_nova = coluna_base + '_BAD_WORD'

        self.banco_dados[nome_col_nova] = (
            self.banco_dados[coluna_base]
            .fillna('')
            .astype(str)
            .apply(lambda v: self._aplicar_regras_bad_word(v, dict_bad_word))
        )

        return self.banco_dados

    def rodar_criticas(self) -> pd.DataFrame:
        self.critica_cep()
        self.critica_uf()
        self.critica_cidade()
        self.critica_bairro()
        return self.banco_dados

    def rodar_bad_word(self, bad_words: list) -> pd.DataFrame:
        self.critica_bad_word(bad_words, 'ENDERECO')
        self.critica_bad_word(bad_words, 'BAIRRO')
        self.critica_bad_word(bad_words, 'CIDADE')
        return self.banco_dados
```

```python
import re
import unicodedata
from num2words import num2words

class NormalizadorEndereco:
    def __init__(self, arquivo_norma="NORMA_ENDERCO.TXT"):
        self.tabela_normalizacao = []

        # Carrega a tabela de normalização
        self.__carregar_tabela(arquivo_norma)

    def __carregar_tabela(self, arquivo):
        """Carrega a tabela de normalização do arquivo"""
        try:
            with open(arquivo, 'r', encoding='utf-8') as f:
                for linha in f:
                    linha = linha.rstrip('\n')
                    if len(linha) >= 40:  # Verifica se a linha tem comprimento mínimo
                        completo = linha[0:20].strip()
                        abrev = linha[20:35].strip()
                        tp_logra = linha[35:36].strip()
                        dups = linha[36:37].strip()
                        tplog_comp = linha[37:38].strip()
                        tipo = linha[38:40].strip()

                        self.tabela_normalizacao.append({
                            'COMPLETO': completo,
                            'ABREV': abrev,
                            'TP_LOGRA': tp_logra,
                            'DUPS': dups,
                            'TPLOG_COMP': tplog_comp,
                            'TIPO': tipo
                        })
        except FileNotFoundError:
            print(f"Arquivo {arquivo} não encontrado")

    # ---------------------- NOVA REGRA DE DEPARA ----------------------

    def aplicar_depara_tabela(self, texto_conteudo: str) -> str:
        """Aplica o depara diretamente percorrendo cada palavra do texto e a tabela de normalização."""
        if not texto_conteudo:
            return texto_conteudo

        texto_conteudo = texto_conteudo.upper()
        palavras = texto_conteudo.split()
        palavras_resultado = []

        for palavra in palavras:
            nova_palavra = palavra

            # Tipo 9 → abreviação para completo
            for registro in self.tabela_normalizacao:
                if registro['TIPO'] == '9':
                    completo = registro['COMPLETO'].upper()
                    abrev = registro['ABREV'].upper()
                    if nova_palavra == abrev:
                        nova_palavra = completo
                        break

            # Tipo 8 → completo para abreviação
            for registro in self.tabela_normalizacao:
                if registro['TIPO'] == '8':
                    completo = registro['COMPLETO'].upper()
                    abrev = registro['ABREV'].upper()
                    if nova_palavra == completo:
                        nova_palavra = abrev
                        break

            # TP_LOGRA == 'S' → logradouros
            for registro in self.tabela_normalizacao:
                if registro['TP_LOGRA'] == 'S':
                    completo = registro['COMPLETO'].upper()
                    abrev = registro['ABREV'].upper()
                    if nova_palavra == completo:
                        nova_palavra = abrev
                        break

            # Tipo 4 → números romanos
            for registro in self.tabela_normalizacao:
                if registro['TIPO'] == '4':
                    completo = registro['COMPLETO'].upper()
                    abrev = registro['ABREV'].upper()
                    if nova_palavra == completo:
                        nova_palavra = abrev
                        break

            # Tipo 2 → preposições (mantém igual)
            for registro in self.tabela_normalizacao:
                if registro['TIPO'] == '2':
                    completo = registro['COMPLETO'].upper()
                    if nova_palavra == completo:
                        nova_palavra = completo
                        break

            # Restante dos casos → completo → abreviação genérica
            for registro in self.tabela_normalizacao:
                tipo = registro['TIPO']
                if tipo not in ['8', '9', '2', '4']:
                    completo = registro['COMPLETO'].upper()
                    abrev = registro['ABREV'].upper()
                    if nova_palavra == completo:
                        nova_palavra = abrev
                        break

            palavras_resultado.append(nova_palavra)

        return ' '.join(palavras_resultado)

    @staticmethod
    def remover_acentos(txt: str) -> str:
        return ''.join(c for c in unicodedata.normalize('NFKD', txt)
                       .encode("ASCII", "ignore").decode("utf-8"))

    def normalizar_nomes(self, texto):
        substituicoes = {
            'FILHO': 'FO',
            'FILH': 'FO',
            'NETO': 'NT',
            'NETTO': 'NT',
            'JUNIOR': 'JR',
            'SOBRINHO': 'SOBR'
        }
        palavras = texto.split()
        resultado = [substituicoes.get(p.upper(), p) for p in palavras]
        return ' '.join(resultado)

    def tratar_sem_numero(self, endereco):
        endereco = re.sub(r'\bS/N\b', 'SN', endereco, flags=re.IGNORECASE)
        endereco = re.sub(r'\b(S|SEM)\s+(NUMERO|N|NR|NRO|NO|NM|NUM)\b',
                          'SN', endereco, flags=re.IGNORECASE)
        return endereco

    def remover_palavras_numero(self, endereco):
        if re.search(r'\d', endereco):
            endereco = re.sub(r'\b(N|NR|NRO|NO|NUMERO|NM|NUM)\s+(?=\d)',
                              '', endereco, flags=re.IGNORECASE)
        return endereco

    def converter_numero_extenso(self, endereco):
        padrao = r'\b(\d{1,2})\s+DE\s+([A-ZÇÃÕÉÊÂÔ]+)\b'

        def repl(m):
            n = int(m.group(1))
            mes = m.group(2)
            extenso = num2words(n, lang='pt_BR')
            extenso = self.remover_acentos(extenso).upper()
            return f'{extenso} DE {mes}'

        return re.sub(padrao, repl, endereco, flags=re.IGNORECASE).upper()

    def tratar_casos_especiais(self, endereco):
        endereco = re.sub(r'\bSOBRE\s+LI\b', 'SLI', endereco, flags=re.IGNORECASE)
        endereco = re.sub(r'\bSOB\s+LI\b', 'SLJ', endereco, flags=re.IGNORECASE)
        endereco = re.sub(r'\b(NS|N|NSA|NOS)\s+SRA\b',
                          'NOSSA SRA', endereco, flags=re.IGNORECASE)
        endereco = re.sub(r'\bCAIXA\s+POSTAL\s+(N|NR|NRO|NO|NUMERO|NM|NUM)\s+',
                          'CAIXA POSTAL ', endereco, flags=re.IGNORECASE)
        return endereco

    def normalizar_endereco(self, endereco_original):
        """Fluxo completo de normalização de endereço"""
        if not endereco_original:
            return ""

        endereco = endereco_original.upper()
        endereco = self.normalizar_nomes(endereco)
        endereco = self.tratar_casos_especiais(endereco)
        endereco = self.converter_numero_extenso(endereco)
        endereco = self.tratar_sem_numero(endereco)
        endereco = self.remover_palavras_numero(endereco)

        # 🚀 Aqui entra sua nova regra de depara linha a linha
        endereco = self.aplicar_depara_tabela(endereco)

        # Remove acentos no final (padronização)
        endereco = self.remover_acentos(endereco)

        # Limpa espaços extras
        endereco = re.sub(r'\s+', ' ', endereco).strip()
        return endereco
```

```python
def gerar_complemento(texto: str):
    if not isinstance(texto, str):
        return None
    # Normaliza NBSP para espaço comum
    t = texto.replace('\u00A0', ' ')
    # Localiza o primeiro número ou "SN"
    m = re.search(r'\b(?:\d+|SN)\b', t, flags=re.IGNORECASE)
    if not m:
        return None
    # Recorta a partir do início do match e normaliza espaços
    parte = t[m.start():]
    parte = re.sub(r'\s+', ' ', parte).strip()
    return parte
```
