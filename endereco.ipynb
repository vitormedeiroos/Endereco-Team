```python
#CLASSIFICANDO TIPO LOGRADOURO SEGUIDO COM ESPACO
#SE COMECAR COM R, RUA COLOCAR RUA
#SE COMECAR COM AV, AVENIDA COLOCAR AVENIDA

regras_logradouro = {
    '^RUA |R ': 'R',
    '^AV |AVENIDA |AVEN ': 'AV',
    '^TRAVESSA |TRAV |TV ': 'TRV',
    '^ALAMEDA |ALAM |ALD ': 'ALAMEDA',
    '^PRAÇA |PCA |PRACA ': 'PRAÇA',
    '^RODOVIA |ROD ': 'RODOVIA',
    '^VIA |V ': 'VIA ',
    '^VIELA |VLE |VL ': 'VIELA',
    '^LARGO |LGO |LG ': 'LARGO',
    'SIT |SITIO |ST ': 'SITIO',
    '^FAZENDA |FAZ |FND ': 'FZND',
    '^CHACARA |CHAC |CH ': 'CHACARA',
    '^VILA |VL ': 'VILA',
    '^\d': 'DIG',
    #em branco
    
}

def aplicar_regras_logradouro(endereco):
    for padrao, substituicao in regras_logradouro.items():
        if re.match(padrao, endereco):
            return substituicao
    return 'OUTROS'

#COLOCAR 
df_endereco_teste['TIPO_LOG'] = df_endereco_teste['ENDERECO'].apply(aplicar_regras_logradouro)
```

### Críticas

```python
class Critica:
    """
    ['ID', 'ENDERECO', 'BAIRRO', 'CIDADE', 'UF', 'CEP']
    """

    def __init__(self, banco_dados: pd.DataFrame):

        self.banco_dados = banco_dados.astype(str)

    @staticmethod
    def _aplicar_regras(valor: str, dict_regras: dict) -> str:

        for padrao, rotulo in dict_regras.items():
            if re.match(padrao, valor):
                return rotulo
        return 'OUTROS'
    
    @staticmethod
    def _aplicar_regras_bad_word(valor: str, dict_regras: dict) -> str:

        for padrao, rotulo in dict_regras.items():
            if re.match(padrao, valor):
                return rotulo
        return 'OK'

    def _extrair_regex(self, nome_coluna_origem: str, nome_coluna_destino: str, dict_regras: dict) -> pd.DataFrame:

        self.banco_dados[nome_coluna_destino] = (
            self.banco_dados[nome_coluna_origem]
            .fillna('')
            .astype(str)
            .apply(lambda v: self._aplicar_regras(v, dict_regras))
        )
        return self.banco_dados

    def critica_cep(self) -> pd.DataFrame:

        self.banco_dados['CEP'] = self.banco_dados['CEP'].str.strip()

        dict_cep = {
            r'^\s*$': 'BRANCO',
            r'^\d{8}$': 'OK',
            r'^(\d)\1+$': 'NÚMERO REPETIDO',
            r'^(?:01234567|12345678|23456789|98765432|87654321|76543210)$': 'NÚMERO SEQUENCIAL',
            r'^(?:012|123|234|345|456|567|678|789|890|987|876|765|654|543|432|321|210)$': 'NÚMERO SEQUENCIAL CURTO'
        }
        return self._extrair_regex('CEP', 'CRITICA_CEP', dict_cep)

    def critica_uf(self) -> pd.DataFrame:

        self.banco_dados['UF'] = self.banco_dados['UF'].str.strip().str.upper()

        dict_uf = {
            r'^\s*$': 'BRANCO',
            r'^[A-Z]{2}$': 'OK'
        }
        return self._extrair_regex('UF', 'CRITICA_UF', dict_uf)

    def _normalizar(self, nome: str):

        # tratar col antes de aplicar reg

        self.banco_dados[nome] = (
            self.banco_dados[nome]
            .str.strip()
            .str.normalize('NFKD')#testar token
            .str.encode(encoding='ascii', errors='ignore') # sem o ignore pode dar erro
            .str.decode('utf-8')
            .str.upper()
        )
        #unicodedata.normalize("NFKD", texto).encode("ASCII", "ignore").decode("utf-8")

    def critica_cidade(self) -> pd.DataFrame:
        
        self._normalizar('CIDADE')

        dict_cidade = {
            r'^\s*$': 'BRANCO',
            r'^[A-Z\s]+$': 'OK',
            r'^[0-9]+$': 'NÚMERO',
            r'^[!@#$%^&*(),.?":{}|<>]+$': 'CARACTERES ESPECIAIS',
            r'^[A-Z\s]*[0-9]+[A-Z\s]*$': 'NÚMERO NO MEIO DO TEXTO',
            r'^[A-Z\s]*[!@#$%^&*(),.?":{}|<>]+[A-Z\s]*$': 'CARACTERES ESPECIAIS NO MEIO DO TEXTO'
        }
        return self._extrair_regex('CIDADE', 'CRITICA_CIDADE', dict_cidade)

    def critica_bairro(self) -> pd.DataFrame:
        self._normalizar('BAIRRO')
        dict_bairro = {
            r'^\s*$': 'BRANCO',
            r'^[A-Z\s]+$': 'OK',
            r'^[0-9]+$': 'NÚMERO',
            r'^[!@#$%^&*(),.?":{}|<>]+$': 'CARACTERES ESPECIAIS',
            r'^[A-Z\s]*[0-9]+[A-Z\s]*$': 'NÚMERO NO MEIO DO TEXTO',
            r'^[A-Z\s]*[!@#$%^&*(),.?":{}|<>]+[A-Z\s]*$': 'CARACTERES ESPECIAIS NO MEIO DO TEXTO'
        }
        return self._extrair_regex('BAIRRO', 'CRITICA_BAIRRO', dict_bairro)


    def critica_bad_word(self, bad_words : list, coluna_base : str) -> pd.DataFrame:

        dict_bad_word = {fr'.*\b{word}\b.*': 'BAD WORD' for word in bad_words}

        nome_col_nova = coluna_base + '_BAD_WORD'

        self.banco_dados[nome_col_nova] = (
            self.banco_dados[coluna_base]
            .fillna('')
            .astype(str)
            .apply(lambda v: self._aplicar_regras_bad_word(v, dict_bad_word))
        )

        return self.banco_dados

    def rodar_criticas(self) -> pd.DataFrame:
        self.critica_cep()
        self.critica_uf()
        self.critica_cidade()
        self.critica_bairro()
        return self.banco_dados

    def rodar_bad_word(self, bad_words: list) -> pd.DataFrame:
        self.critica_bad_word(bad_words, 'ENDERECO')
        self.critica_bad_word(bad_words, 'BAIRRO')
        self.critica_bad_word(bad_words, 'CIDADE')
        return self.banco_dados
```

```python
class NormalizadorEndereco:
    def __init__(self, arquivo_norma="NORMA_ENDERCO.TXT"):
        self.tabela_normalizacao = []
        self.indices = {}

        # Carrega a tabela de normalização
        self.__carregar_tabela(arquivo_norma)
        self.__criar_indices()
 

    def __carregar_tabela(self, arquivo):
        """Carrega a tabela de normalização do arquivo"""
        try:
            with open(arquivo, 'r', encoding='utf-8') as f:
                for linha in f:
                    linha = linha.rstrip('\n')
                    if len(linha) >= 40:  # Verifica se a linha tem comprimento mínimo
                        completo = linha[0:20].strip()
                        abrev = linha[20:35].strip()
                        tp_logra = linha[35:36].strip()
                        dups = linha[36:37].strip()
                        tplog_comp = linha[37:38].strip()
                        tipo = linha[38:40].strip()
 
                        self.tabela_normalizacao.append({
                            'COMPLETO': completo,
                            'ABREV': abrev,
                            'TP_LOGRA': tp_logra,
                            'DUPS': dups,
                            'TPLOG_COMP': tplog_comp,
                            'TIPO': tipo
                        })
            
        except FileNotFoundError:
            print(f"Arquivo {arquivo} não encontrado")
 
    def __criar_indices(self):
        """Cria índices para busca rápida"""
        self.indices = {
            'completo_para_abrev': {},
            'abrev_para_completo': {},
            'tipos_logradouro': set(),
            'preposicoes': set(),
            'numeros_romanos': {},
            'titulos': set()
        }
 
        for registro in self.tabela_normalizacao:
            completo = registro['COMPLETO']
            abrev = registro['ABREV']
            tipo = registro['TIPO']
 
            # Índice completo -> abreviação
            if completo:
                self.indices['completo_para_abrev'][completo.upper()] = abrev
 
            # Índice abreviação -> completo (para alguns casos)
            if abrev and tipo in ['1', '3']:
                self.indices['abrev_para_completo'][abrev.upper()] = completo
 
            # Tipos de logradouro
            if registro['TP_LOGRA'] == 'S':
                self.indices['tipos_logradouro'].add(completo.upper())
                if abrev:
                    self.indices['tipos_logradouro'].add(abrev.upper())
 
            # Preposições
            if tipo == '2':
                self.indices['preposicoes'].add(completo.upper())
 
            # Números romanos
            if tipo == '4':
                self.indices['numeros_romanos'][completo.upper()] = abrev
 
            # Títulos
            if tipo in ['8', '9']:
                self.indices['titulos'].add(completo.upper())
                if abrev:
                    self.indices['titulos'].add(abrev.upper())
 
    @staticmethod
    def remover_acentos(txt: str) -> str:
        return ''.join(c for c in unicodedata.normalize('NFKD', txt).encode("ASCII", "ignore").decode("utf-8"))
    
    def normalizar_nomes(self, texto):
        """Aplica normalização de nomes (FILHO, NETO, etc.)"""
        substituicoes = {
            'FILHO': 'FO',
            'FILH': 'FO',
            'NETO': 'NT',
            'NETTO': 'NT',
            'JUNIOR': 'JR',
            'SOBRINHO': 'SOBR'
        }
 
        palavras = texto.split()
        resultado = []
 
        for palavra in palavras:
            palavra_upper = palavra.upper()
            if palavra_upper in substituicoes:
                resultado.append(substituicoes[palavra_upper])
            else:
                resultado.append(palavra)
 
        return ' '.join(resultado)
 
    def tratar_sem_numero(self, endereco):
        """Substitui 'S/N' e variações por 'SN'"""
        padrao_s_n = r'\b(S|SEM)\s+(NUMERO|N|NR|NRO|NO|NM|NUM)\b'
        padrao_sn_simples = r'\bS/N\b'
 
        # Substitui S/N simples
        endereco = re.sub(padrao_sn_simples, 'SN', endereco, flags=re.IGNORECASE)
 
        # Substitui combinações como "SEM NUMERO"
        endereco = re.sub(padrao_s_n, 'SN', endereco, flags=re.IGNORECASE)
 
        return endereco
 
    def remover_palavras_numero(self, endereco):
        """Remove palavras como N, NR, NRO após encontrar número"""
        # Primeiro, identifica se há número no endereço
        if re.search(r'\d', endereco):
            padrao_palavras_numero = r'\b(N|NR|NRO|NO|NUMERO|NM|NUM)\s+(?=\d)'
            endereco = re.sub(padrao_palavras_numero, '', endereco, flags=re.IGNORECASE)
 
        return endereco    

    def converter_numero_extenso(self, endereco):
        """Converte números por extenso em casos como '25 DE SETEMBRO'"""
        # Busca por padrão "NÚMERO DE MÊS"
        padrao = r'\b(\d{1,2})\s+DE\s+([A-ZÇÃÕÉÊÂÔ]+)\b'

        def repl(m):
            n = int(m.group(1))
            mes = m.group(2)            
            extenso = num2words(n, lang='pt_BR')
            extenso = self.remover_acentos(extenso).upper()
            return f'{extenso} DE {mes}'

        return re.sub(padrao, repl, endereco, flags=re.IGNORECASE).upper()
 
    def tratar_casos_especiais(self, endereco):
        """Aplica casos especiais não incluídos na tabela"""
        # SOBRE LI -> SLI
        endereco = re.sub(r'\bSOBRE\s+LI\b', 'SLI', endereco, flags=re.IGNORECASE)
 
        # SOB LI -> SLJ
        endereco = re.sub(r'\bSOB\s+LI\b', 'SLJ', endereco, flags=re.IGNORECASE)
 
        # NS SRA -> NOSSA SRA
        endereco = re.sub(r'\b(NS|N|NSA|NOS)\s+SRA\b', 'NOSSA SRA', endereco, flags=re.IGNORECASE)
 
        # CAIXA POSTAL NUM -> CAIXA POSTAL
        endereco = re.sub(r'\bCAIXA\s+POSTAL\s+(N|NR|NRO|NO|NUMERO|NM|NUM)\s+', 'CAIXA POSTAL ', endereco,
                          flags=re.IGNORECASE)
 
        return endereco
 
    def normalizar_palavra(self, palavra):
        """Normaliza uma palavra individual usando a tabela"""
        palavra_upper = palavra.upper()
 
        # Verifica se a palavra está no índice de completo para abreviação
        if palavra_upper in self.indices['completo_para_abrev']:
            return self.indices['completo_para_abrev'][palavra_upper]
 
        # Se não encontrou, retorna a palavra original
        return palavra
 
    def normalizar_endereco(self, endereco_original):
        """Função principal que normaliza um endereço completo"""
        if not endereco_original:
            return ""
 
        # Aplica normalização de nomes primeiro
        endereco = self.normalizar_nomes(endereco_original)
 
        # Aplica casos especiais
        endereco = self.tratar_casos_especiais(endereco)
 
        # Converte números por extenso
        endereco = self.converter_numero_extenso(endereco)
 
        # Trata "sem número"
        endereco = self.tratar_sem_numero(endereco)
 
        # Remove palavras de número desnecessárias
        endereco = self.remover_palavras_numero(endereco)
 
        # Divide o endereço em palavras e normaliza cada uma
        palavras = endereco.split()
        palavras_normalizadas = []
 
        for palavra in palavras:
            # Mantém números e SN intactos
            if palavra.isdigit() or palavra.upper() == 'SN':
                palavras_normalizadas.append(palavra)
            else:
                palavra_normalizada = self.normalizar_palavra(palavra)
                palavras_normalizadas.append(palavra_normalizada)
 
        # Reconstroi o endereço
        endereco_normalizado = ' '.join(palavras_normalizadas)
 
        # Limpa espaços extras
        endereco_normalizado = re.sub(r'\s+', ' ', endereco_normalizado).strip()
 
        return endereco_normalizado
```
